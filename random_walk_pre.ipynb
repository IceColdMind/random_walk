```julia
function line_random_walk(n)
    current=0
    for i=1:n
        rng=rand(1:2)
        if rng==1
            current +=1
        else
            current -=1
        end 
    end
    return current
end
```

```julia
function loop_through_random_walk(func,steps,times)
    array=[]
    for i=1:times
        array=push!(array,func(steps))
    end
    return array
end
```

```julia
using Plots
```

```julia
arr=loop_through_random_walk(line_random_walk,1000,50000)
```
